```python
import random


# Ask if the user's will to play game theory
user_will=input(prompt="Do you want to play a Game Theory situation?(Answer YES or No)").lower()

#error check of incorrect input
while("yes" not in user_will and "no" not in user_will):
    user_will=input(prompt="\tInput is invalid.Please input YES or No)").lower()
    
#If user want to play    
if(user_will=="yes"):
    
# Board Game Rule Explanation
    user_name=input(prompt="Hi,welcome to this game! You will fight with me! May I have your name?").capitalize()
    print(f"\n{user_name} we are two business competitors. We are now thinking pricing strategy for our service!")
    print("""
    You are an marketing directer at companyA. You can't communicate with anyone outside companyA.
    According to market research, our profits will change as follows :""")
    print(f"""
    {user_name}, this is your situation, think well before answering.
    \n
    {user_name} \ Computer\t\t Keep price \t\t\t  Decrease price
    --------------\t --------------------------\t ---------------------------
    Keep  price\t\t Both gets $100\t\t\t {user_name} gets $70, Computer gets $110
    Decrease price \t {user_name} gets $110,Computer gets $70  Both gets $90
    """)
    #Score Chart of Board Game
    board_table=[[90,110],[70,100]]
    
    #Dictionary of two player's choice to allocate score R: keep price, D: decrease price
    result_dic={"(D,D)":(board_table[0][0],board_table[0][0]),
                "(D,K)":(board_table[0][1],board_table[1][0]),
                "(K,D)":(board_table[1][0],board_table[0][1]),
                "(K,K)":(board_table[1][1],board_table[1][1])}
    
    # ask the user to select what to do    
    user_choice=input(prompt="""
    What would you do? Would you decrease price or keep price?
    Select:
    - 'D' to decrease price
    - 'K' to keep price
    """).upper()
    
    #error check of incorrect input
    while(user_choice!="D" and user_choice!="K"):
        user_choice=input(prompt="\tInput is invalid. Input D or K").upper()
        
    # Computer randomly either to keep price or decrease price
    computer_choice=random.randint(0,1)
    
    # conditional structure with the outputs of each scenario
    if user_choice=="D" and computer_choice==0:
        user_result=result_dic["(D,D)"][0]
        computer_result=result_dic["(D,D)"][1]
        print(f"""
        You and me decided to decrease price!
        You earned {user_result} points.
        I earned {computer_result} points""")
    
    elif user_choice=="K" and computer_choice==0:
        user_result=result_dic["(K,D)"][0]
        computer_result=result_dic["(K,D)"][1]
        print(f"""
        You decided to keep price, I decreased!
        You earned {user_result} points.
        I earned {computer_result} points""")
    
    elif user_choice=="D" and computer_choice==1:
        user_result=result_dic["(D,K)"][0]
        computer_result=result_dic["(D,K)"][1]
        print(f"""
        I decided to keep price, you decreased!
        You earned {user_result} points.
        I earned {computer_result} points""")
        
    elif user_choice == "K" and computer_choice == 1:
        user_result = result_dic["(K,K)"][0]
        computer_result = result_dic["(K,K)"][1]
        print(f"""
        We both kept price.
        You earned {user_result} points.
        I earned {computer_result} points""")
    
    # Declaring a winner.
    if user_result > computer_result:
        print(f"""Congrats, you won!""")

    elif user_result == computer_result:
        print(f"""Well we tie!""")

    elif user_result <= computer_result:
        print(f"""\tCongrats to me, I beat you!""")
        
#If user do not want to play 
else:
    
# Board Game Rule Explanation
    print("""
    User1 and User2 are marketing directers at companyA and companyB!
    They are now thinking about new pricing strategy!""")
    print(f"""
    This is their situation.
    \n
    User1 \ User2\t\t Keep price \t\t\t  Decrease price
    --------------\t --------------------------\t ---------------------------
    Keep price\t\t Both gets $100\t\t\t User1 gets $70, user2 gets $110
    Decrease price \t User1 gets $110,user2 gets $70  Both gets $90
    """)
    
    #Score Chart of Board Game
    board_table=[[90,110],[70,100]]
    
    #Dictionary of two player's choice to allocate score R: keep price, D: decrease price
    result_dic={"(D,D)":(board_table[0][0],board_table[0][0]),
                "(D,K)":(board_table[0][1],board_table[1][0]),
                "(K,D)":(board_table[1][0],board_table[0][1]),
                "(K,K)":(board_table[1][1],board_table[1][1])}
    
    #Set the list to store the each user's point in each round
    pointOfuser1=[]
    pointOfuser2=[]
    
    #set variables to store total scores of each user in five rounds
    user1_total,user2_total=0,0
    
    #Initialize a valuable to store whether or not user1 kept price in the previous round
    Was_user1_Snitch=False
    
    #Conduct the game five times
    for i in range(0,5):
        
        # Computer1 randomly either to keep price or decrease price
        user1_choice=random.randint(0,1)
        
         # Computer2 choose to keep silent or betray based on whether or not Computer1 kept price
         # Computer2 always decrease price at first round,and follow the computer1's strategy in following rounds
        if(Was_user1_Snitch==False):
            user2_choice="D"
        else:
            user2_choice="K"
        
        # Conditional structure with the outputs of each scenario
        if user1_choice==0 and user2_choice=="D":
            user1_result=result_dic["(D,D)"][0]
            user2_result=result_dic["(D,D)"][1]
            print(f"""
            User1 and user2 decided to decrease price!
            User1 earned {user1_result} points.
            User2 earned {user2_result} points""")
            
            # Regard computer1 as not snitch
            Was_user1_Snitch=False
    
        elif user1_choice==1 and user2_choice=="D":
            user1_result=result_dic["(K,D)"][0]
            user2_result=result_dic["(K,D)"][1]
            print(f"""
            User1 decided to keep price, user2 decreased!
            User1 earned {user1_result} points.
            User2 earned {user2_result} points""")
            
            # Regard computer1 as snitch
            Was_user1_Snitch=True
    
        elif user1_choice==0 and user2_choice=="K":
            user1_result=result_dic["(D,K)"][0]
            user2_result=result_dic["(D,K)"][1]
            print(f"""
            User2 decided to keep price, user1 decreased!
            User1 earned {user1_result} points.
            User2 earned {user2_result} points""")
            
            # Regard computer1 as not snitch
            Was_user1_Snitch=False
        
        elif user1_choice == 1 and user2_choice == "K":
            user1_result = result_dic["(K,K)"][0]
            user2_result = result_dic["(K,K)"][1]
            print(f"""
            both user1 and user2 kept price.
            User1 earned {user1_result} points.
            User2 earned {user2_result} points""")

            # Regard computer1 as snitch
            Was_user1_Snitch=True
        
        #Display the result of each round
        if user1_result > user2_result:
            print(f"""
            In Round{i+1},user1 won!""")

        elif user1_result == user2_result:
            print(f"""
            In Round{i+1},they tie!""")

        elif user1_result <= user2_result:
            print(f"""
            In Round{i+1},user2 won!""")
            
        #Store each user's point in one round
        pointOfuser1.append(user1_result)
        pointOfuser2.append(user2_result)
    
    #Aggregate each user's scores 
    for i in pointOfuser1:  
        user1_total+=i
    for j in pointOfuser2:
        user2_total+=j
    
    # Declaring a winner. 
    if(user1_total>user2_total):
        print("\t    Finally,winner is user1")
    elif(user1_total<user2_total):
        print("\t    Finally,winner is user2")
    else:
        print("\t    Finally,they tie")
    
```
